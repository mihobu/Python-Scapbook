## How to plot an empirical CDF

```python
# Get some randomly distributed values
s = np.random.lognormal(mean=0.5, sigma=0.75, size=100)
```

```python
# Sort
x = np.sort(s)

# Calculate cumulative distribution
y = np.arange(len(x)) / float(len(x))
```

```python
plt.figure(figsize=(8,8), dpi=100)
plt.plot(x, y, marker='')
#plt.plot([0,1], [0,1], c='red') # red reference line
#plt.xticks(np.arange(0,1.1,0.1))
#plt.yticks(np.arange(0,1.1,0.1))
#plt.xlabel('False Positive Rate (FPR)')
#plt.ylabel('True Positive Rate (TPR)')
plt.grid(linestyle='dotted')
#plt.axis(xmin=-0.05, xmax=1.05, ymin=-0.05, ymax=1.05)
#plt.plot(bestX, bestY, 'r*', markersize=20)
plt.show()
```
